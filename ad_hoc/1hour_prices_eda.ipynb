```python
import pyspark.sql.functions as sf
from time import time
spark.conf.set("spark.sql.session.timeZone", "America/New_York")
```

```python
# How mmuch time should be analyzed in seconds
window_time = time() - 100000
```

```python
# Read data from 'runescape.02_silver.item_mapping'
df_map = spark.read.table("runescape.02_silver.item_mapping")
```

```python
# Read data from 'runescape.01_bronze.1h_prices' to get Volumes
df_1hour = spark.read.table("runescape.01_bronze.1h_prices")
# Filter to only use the most recent Volume data
df_1hour = df_1hour.filter(f"time >= {window_time}")
df_1hour = df_1hour.join(df_map,"id")
df_1hour = df_1hour.sort("time")
#df_1hour = df_1hour.withColumn("profit", sf.round((sf.col("highalch") - sf.col("avg_price")) - 100,1))
```

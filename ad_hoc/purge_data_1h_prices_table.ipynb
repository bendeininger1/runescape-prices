Purge old data in 'runescape.01_bronze.1h_prices' 

```python
from pyspark.sql.functions import *
from time import time
```

```python
# Read data from 'runescape.01_bronze.1h_prices'
df = spark.read.table("runescape.01_bronze.1h_prices")
print(df.count())
df = df.filter("time > 1770099593") # exclude older data
print(df.count())
```

Overwrite data

```python
# Write the DataFrame to a Unity Catalog managed Delta table in the bronze schema, overwriting the table
df.write.mode("overwrite").saveAsTable("runescape.01_bronze.1h_prices")
```

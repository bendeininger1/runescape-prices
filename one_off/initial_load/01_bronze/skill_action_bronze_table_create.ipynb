This notebook creates the table 01_bronze.skill_action_inputs used to store skill action inputs and outputs.  
This notebook creates the table 01_bronze.skill_action_xp used to store skill action xp.  

```python
from pyspark.sql.types import StructType, StructField, LongType, StringType, IntegerType, DoubleType
```

```python
catalog = dbutils.widgets.get("catalog")
```

```python
schema_skill_action_inputs = StructType(
    [
        StructField("output", StringType(), False),
        StructField("input_qty", DoubleType(), False),
        StructField("input", StringType(), False),
        StructField("skill", StringType(), False)
    ]
)
```

```python
schema_skill_action_xp = StructType(
    [
        StructField("output", StringType(), False),
        StructField("xp", DoubleType(), False),
        StructField("skill", StringType(), False)
    ]
)
```

```python
if not spark.catalog.tableExists(f"{catalog}.01_bronze.skill_action_inputs"):
    spark.catalog.createTable(f"{catalog}.01_bronze.skill_action_inputs", schema=schema_skill_action_inputs)
```

```python
if not spark.catalog.tableExists(f"{catalog}.01_bronze.skill_action_xp"):
    spark.catalog.createTable(f"{catalog}.01_bronze.skill_action_xp", schema=schema_skill_action_xp)
```

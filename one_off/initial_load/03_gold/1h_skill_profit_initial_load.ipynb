This notebook creates the table 1h_skill_profit in the gold layer used to store skill action profits.

```python
from pyspark.sql.types import StructType, StructField, LongType, StringType, IntegerType, DoubleType
```

```python
schema = StructType(
    [
        StructField("time", LongType(), False),
        StructField("output", StringType(), False),
        StructField("skill", StringType(), False),
        StructField("output_avg_high", IntegerType(), False),
        StructField("output_high_volume", IntegerType(), False),
        StructField("output_avg_low", IntegerType(), False),
        StructField("output_low_volume", IntegerType(), False),
        StructField("id_output", LongType(), False),
        StructField("total_input_cost_high", DoubleType(), False),
        StructField("total_input_cost_low", DoubleType(), False),
        StructField("profit_low_est", DoubleType(), True),
        StructField("profit_high_est", DoubleType(), True),
    ]
)
```

```python
spark.catalog.createTable("runescape.03_gold.1h_skill_profit", schema=schema)
```

This notebook creates the table 02_silver.latest_prices_enriched used to store cleansed price data every minute.

```python
from pyspark.sql.types import StructType, StructField, LongType, StringType, IntegerType, DoubleType
```

```python
catalog = dbutils.widgets.get("catalog")
```

```python
schema = StructType(
    [
        StructField("id", IntegerType(), False),
        StructField("price", IntegerType(), False),
        StructField("time", LongType(), False),
        StructField("highorlow", StringType(), False),
        StructField("name", StringType(), False),
        StructField("highalch", IntegerType(), True),
        StructField("limit", IntegerType(), True),
        StructField("members", StringType(), False)
    ]
)
```

```python
if not spark.catalog.tableExists(f"{catalog}.02_silver.latest_prices_enriched"):
    spark.catalog.createTable(f"{catalog}.02_silver.latest_prices_enriched", schema=schema)
```

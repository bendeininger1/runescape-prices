```python
# Read data from bronze layer Unity Catalog table 'runescape.01_bronze.shop_prices'
df_shop_prices = spark.read.table("runescape.01_bronze.shop_prices")
```

```python
# Read data from silver layer mapping table 'runescape.02_silver.item_mapping'
df_map = spark.read.table("runescape.02_silver.item_mapping").\
    select("name", "id")
```

```python
# join data frames
df = df_shop_prices.join(df_map, "name")
```

```python
# filter out low volume crap
df = df.filter(df.id.isin(
    [1609,
     1611,
     1613,
     1625,
     1627,
     1629,
     6311,
     6313,
     6315,
     6317]))
```

```python
df.display()
```

```python
# Write data to silver layer Unity Catalog table 'runescape.02_silver.shop_prices_enriched'
df.write.mode("overwrite").saveAsTable("runescape.02_silver.shop_prices_enriched")
```

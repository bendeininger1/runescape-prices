This notebook creates the table 1h_prices_latest_enriched

```python
import pyspark.sql.functions as sf
```

```python
# Read data from 'runescape.02_silver.1h_prices_cleansed'
df_1hour_cleansed = spark.read.table("runescape.02_silver.1h_prices_cleansed")
```

```python
# create df with only the latest 1h_prices
# get latest time for each id
df_latest_time = df_1hour_cleansed.groupBy("id").max("time").withColumnRenamed("max(time)", "time")
df_latest = df_1hour_cleansed.join(df_latest_time, ["id", "time"])
```

This drops all tables for dev/ testing purposes

```python
# Specify your catalog and schema
catalog = dbutils.widgets.get("catalog")
if catalog == "runescape_prod":
    raise Exception ("You probably dont want to delete all tables in prod")

schema = ["01_bronze","02_silver","03_gold"]

for schema in schema:
    # Get all table names
    tables = spark.sql(f"SHOW TABLES IN {catalog}.{schema}")

    # Loop and drop each table
    for row in tables.collect():
        table_name = row['tableName']
        spark.sql(f"DROP TABLE IF EXISTS {catalog}.{schema}.{table_name}")
```

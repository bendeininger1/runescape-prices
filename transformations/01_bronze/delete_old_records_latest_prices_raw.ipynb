Delete old records in the table 'runescape.01_bronze.latest_prices_raw' to improve performance  
Called by job

```python
import pyspark.sql.functions as sf
```

```python
# Read data from 'runescape.01_bronze.latest_prices_raw'
df = spark.read.table("runescape.01_bronze.latest_prices_raw")
print(df.count())
```

```python
# How far back should data be deleted from? in seconds
time_to_delete = 86400
df = df.filter((sf.col("time") > sf.unix_timestamp() - time_to_delete))
print(df.count())
```

```python
# Write the DataFrame to a Unity Catalog managed Delta table in the bronze schema, overwritting the data
df.write.mode("overwrite").saveAsTable("runescape.01_bronze.latest_prices_raw")
```

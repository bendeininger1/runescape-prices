This notebook archives "old" 1m price json file moving them from /Volumes/{catalog}/00_landing/data_sources/latest_prices to /Volumes/{catalog}/00_landing/data_sources/1m_prices_archive.  This job only moves files that have likely already been processed into the bronze layer by only moving files with time stamps greater than 5 minutes old.

```python
from time import time
import urllib.request
import os
import shutil
```

```python
catalog = dbutils.widgets.get("catalog")
```

```python
# Define and create the local directory where the files will be written to
dest_path = f"/Volumes/{catalog}/00_landing/data_sources/1m_prices_archive"
os.makedirs(dest_path, exist_ok=True)


# Define the full path for the downloaded data file
# TODO update directory to f"/Volumes/{catalog}/00_landing/data_sources/1m_prices/"
source_path = f"/Volumes/{catalog}/00_landing/data_sources/latest_prices/"

# get unix time for checking files
unix_timestamp = int(time())

# set threshould of 5 minutes for determing which files are "old"
time_filter = unix_timestamp - 5*60
```

```python
# TODO move to utils
# Loop over files in source_path and move the files if the meet the "age" criteria
directory = os.fsencode(source_path)
    
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    print (filename)
    if int(filename[-15:-5])<time_filter: 
        # print(os.path.join(directory, filename))
        shutil.move(source_path+filename, dest_path) 
        continue
```
